Appending to a File


```python
# Note if we dont put an extra line at the end of our txt file then the Text will be added whitout a new line like:

# Garry - DirectorAsh - Human Resources 
```

```python
employee_file = open("27_employees.txt", "a") # a == append

employee_file.write("Ash - Human Resources") # Will add a new employee at the end of the file, Why at end? cuz we are using append 
    
employee_file.close()
```

```python
# How to not mess up our file is to add a new line before we enter out text like below:

employee_file = open("27_employees.txt", "a") # a == append

employee_file.write("\nKelly - Customer Service")
    
employee_file.close()
```

Overwriting Files

```python
employee_file = open("27_employees_Overwrite.txt", "w") # w == overwrite

employee_file.write("Kelly - Customer Service")
    
employee_file.close()
```

```python
# If we dont have a file to overwrite then it creates a files and writes in it

employee_file = open("index.html", "w") # w == overwrite

employee_file.write("<b>This is a Button</b>")
    
employee_file.close()
```
