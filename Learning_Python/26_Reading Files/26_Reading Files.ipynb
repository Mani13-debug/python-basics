```python
employee_file = open("26_employees.txt", "r")
print(employee_file.readable()) # Tells if the file is readable or not

employee_file.close() # Closes the open file
```

```python
employee_file = open("26_employees.txt", "w") # w means we cannot read it but we can write in it
print(employee_file.readable()) 

employee_file.close() # Note when we run the above code it deletes the contents of the file!
```

```python
# Gives the information in file
employee_file = open("26_employees.txt", "r")
print(employee_file.read()) 

employee_file.close()
```

Reading Lines One by One

```python
# Printing the lines in the fileone by one
employee_file = open("26_employees.txt", "r")
print(employee_file.readline())

employee_file.close()
```

```python
employee_file = open("26_employees.txt", "r")
print(employee_file.readline())
print(employee_file.readline())
print(employee_file.readline())

employee_file.close()
```

Reading Multiple lines at once

```python
employee_file = open("26_employees.txt", "r")
print(employee_file.readlines()) # Prints the contents a list
    

employee_file.close()
```

```python
employee_file = open("26_employees.txt", "r")
print(employee_file.readlines()[1]) # Prints the contents a list
    

employee_file.close()
```

```python
employee_file = open("26_employees.txt", "r")
for employee in employee_file.readlines():
    print(employee)

employee_file.close()
```
