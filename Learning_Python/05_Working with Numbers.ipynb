```python
# pow --> power function
print(pow(3 , 2)) # can also be written as "3 ** 2"
print(pow(333 , 4))

#max --> maximum of two numbers
print(max(333 , 4)) 

#min --> minimum of two numbers
print(min(333 , 4))

# round off the given number
print(round(333.0923411789))
```

```python
from math import *

print(floor(333.0923411789)) # removes the decimals and the returns the same num
print(ceil(333.0923411789)) # removes te decimals and does num + 1
print(sqrt(333.0923411789)) # sqrt --> square root
```
