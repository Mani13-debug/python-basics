### Adding a Removing elements in a List

##### Adding in Lists

```python
numbers = [7, 13, 42, 101, 1001, 10]
friends = ["Kevin", "Karen", "Jim", "Oscar", "Tim", "Jim"]

print(numbers)
print(friends)
```

```python
# Takes a list and appends the elements of the list mentioned in another list
friends.extend (numbers)
print(friends)
```

```python
# Adds elements at the end of the list
friends.append ("Creed")
print(friends)
```

```python
# Adds at what index you want to add
friends.insert (1, "Macho")
print(friends)

# format = list.insert(index, what you want to add)
```

##### Removing in Lists

```python
numbers = [7, 13, 42, 101, 1001, 10]
friends = ["Kevin", "Karen", "Jim", "Oscar", "Tim", "Jim"]
```

```python
# Removing Items from a list
# format = list.remove(element to remove)

numbers.remove(13)
print(numbers)
```

```python
# Deleting a whole list

numbers.clear()
print(numbers)
```

```python
# Deleting a elemnt at a particular Index
# if no index is provided then it deletes the elemnt at the end of the list

friends.pop(3)
print(friends)
```

#### Reversing a List

```python
numbers = [7, 13, 42, 101, 1001, 10]
friends = ["Kevin", "Karen", "Jim", "Tim"]

# Method 1
numbers.reverse()
print(numbers)

friends.reverse()
print(friends)

# Note print(numbers.reverse()) is wrong!!
```

```python
numbers = [7, 13, 42, 101, 1001, 10]
friends = ["Kevin", "Karen", "Jim", "Tim"]

# Method 2

print(numbers[::-1])
print(friends[::-1])
```

#### Sorting elements in a List

```python
numbers = [7, 13, 42, 101, 1001, 10]
friends = ["Kevin", "Karen", "Jim","Tim"]
```

```python
# Doing this is not the right approach THis wikk give output as None
sort1 = friends.sort() 
print(sort1)

# The right way

# Sort arranges in ascending way
numbers.sort()
friends.sort()

print(numbers)
print(friends)
```

```python
# Sorting in descending order

numbers.sort(reverse = True)
friends.sort(reverse = True) 

print(numbers)
print(friends)
```

```python
# Sorted Function

print(sorted(numbers)) # Ascending Order
print(sorted(numbers, reverse = True)) # Descending Order
```

#### Misleneous

```python
numbers = [7, 13, 42, 101, 1001, 10]
friends = ["Kevin", "Karen", "Jim", "Tim", "Oscar", "Tim"]
```

```python
# Returns the index of the particular element mentioned
# Format = list.inex(element)

print(friends.index("Oscar"))
print(numbers.index(10))
```

```python
# Counting how many times a element comes in a list

print(friends.count("Tim"))
```

```python
# Copying a list

friends2 = friends.copy()
print(friends2)
```
