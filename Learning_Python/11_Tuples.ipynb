```python
# A tuple is created by writing in ()
# A tuple cannot be changed once it is made i.e it is immutable.
```

```python
coordinates = (4, 5)
# Index        0  1
print(coordinates)
print(coordinates[0])
```

```python
coordinates[1] = 10 # Not changable
```

#### Indexing and Slicing in Tuples

```python
# Tuples inside a list
coordinates = [(4, 5), (6, 7), (40, 34)]
# Index         0 1     0  1     0   1
# Index          0        1        2
print(coordinates[0])
print(coordinates[1])
```

```python
# Nested tuples inside a list

print(coordinates[0][1])
print(coordinates[2][0])
```

```python
# Other indexing and Slicing methods are same as Lists
```

#### Unpacking and Packing of Tuples

```python
# Packing means assigning values to a tuple
# Unpacking means extracting values from a tuple and put them in variables
```

```python
cars = ("Audi", "Mercedes", "BMW")
car1, car2, car3  =  cars
#---Variables---     Tuple
print(car1)
print(car2)
print(car3)
```

```python
# Use of Asterist in unpacking atuple
# Used when th number of variables are less than the values of a tuple.

cars = ("Audi", "Mercedes", "BMW", "Toyota", "Ford")

car1, car2, *car3  =  cars
# If asterisk is not used it will give a Value error
print(car1)
print(car2)
print(car3)
```

```python
# If asterisk is used other than last variable

cars = ("Audi", "Mercedes", "BMW", "Toyota", "Ford")

car1, *car2, car3  =  cars
# Two asterisk cannot be used! 
print(car1)
print(car2)
print(car3)
```
