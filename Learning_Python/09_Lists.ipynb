#### Indexing in Lists

```python
# List can have all types of data types such as Integer, Boolean, and String

animes = ["One Piece", "Death Note" , "One Punch Man", "Spirit Chronicles" , "Mushoku Tensei" , "AOT"]
#  Index      0              1               2                    3                  4            5
#  Index     -6             -6              -4                    -3                   -2        -1      
print(animes)
```

```python
# Positive indexing helps us to read/print a list from the start part of the list
print(animes[0])
print(animes[2])
print(animes[1])

# Negative indexing helps us read/print from end part of the list
print(animes[-1]) 
print(animes[-2])
print(animes[-3])
```

#### Slicing in lists

```python
# General Format

# list[start:end:step]
```

```python
print(animes[:4])
print(animes[1:])
```

```python
print(animes[1:4])
print(animes[1:-3])
```

```python
print(animes[-3:-3])
```

```python
# What does Step do?
# step argument determines the interval at which elements are selected from the list.

print(animes[1:4:2])
print(animes[0:6:3])
```

```python
num = [0, 1, 2, 3, 4, 5, 6, 7, 8]

print(num[0:6:2]) # Will pring evn numbers from index position 0 to 5
print(num[2:9:3])

"""
One thing I observed !!!
When step arguement = 2 it does not mean it will skip two elements it will only skip 1
simillarly if split arguement = 3 it means it will skip 2 elements
"""
# i.e this means it will skip (n-1) elements
```

```python
print(num[::2]) # takes every 2nd element 
print(num[::-2])
print(num[0:4:-1]) # Does not Make sense
print(num[-8:-4:-3]) # Does not Make sense

# Also ::2 --> 0:9:2
# also element on the 9th index will not be considered
```

#### List Comprehensions

```python
# List Comprehensions = A consice way to create lists in Python
#                       compact and easier to read than the traditional loops
#                       [expression for value in iterable if condition] 
```

```python
# Normal way

doubles = []
for x in range(1,11):
  doubles.append(x * 2)
print(doubles)
```

```python
# Comprehension

doubles = [x * 2 for x in range(1,11)]
print(doubles)
```

```python
triples = [y * 3 for y in range(1,11)]
print(triples)
```

```python
fruits = ['apple', 'orange', 'banana', 'coconut']
fruits_up = [fruit.upper() for fruit in fruits]
print(fruits_up)
```

```python
fruit_chars = [fruit[0] for fruit in fruits]
print(fruit_chars)
```

```python
numbers = [1, -2, 3, -4, 5, -6]
positive_nums = [num for num in numbers if num >= 0]
print(positive_nums)

negative_num = [num for num in numbers if num < 0]
print(negative_num)

even_numbers = [num for num in numbers if num % 2 == 0]
print(even_numbers)
```

```python
grades = [85, 42, 79, 90, 56, 61, 30]
passing_grades = [grade for grade in grades if grade >= 60]

print(passing_grades)
```

#### Nested Lists

```python
nested_list = [[10, 20, 30], [44, 55, 66], [77, 87, 99]]
# Index          0   1   2     0   1   2     0   1   2
# Index              0             1             2

print(nested_list[1][1])
```

```python
list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]

list1[2][2].append(7000)
print(list1)
```

```python
list1 = ["a", "b", ["c", ["d", "e", ["f", "g"], "k"], "l"], "m", "n"]

# sub list to add
sub_list = ["h", "i", "j"]

list1[2][1][2].extend(sub_list)
print(list1)
```
