#### If statements with comparison operators

```python
# ==, !=, >, <, >=, <=  --> comparison operators

# == means equal to
# != means not equal to
# > means greater than
# < means less than
# >= means greater than equal to
# <= means less than equal to
```

```python
num1 = int(input("Enter num1: "))
num2 = int(input("Enter num2: "))
num3 = int(input("Enter num3: "))

if num1 >= num2 and num1 >= num3:
  print(num1)
elif num2 >= num1 and num2 >= num3:
    print(num2)
else:
    print(num3)
```

#### Logical Operators 

```python
# logical operators = used on conditional statements

#   and = checks two or more conditions if True
#   or = checks if atleast one condition is true
#   not = True if condition is False, and vice versa
```

```python
temp = 25

if temp > 0 and temp < 30: # and --> if one of the condition is false that means the whole statement itself is false
  print("The temperature is good.")
else:
  print("The temperature is bad.")
```

```python
temp = 40

if temp <= 0 or temp >= 30: # or --> if one of the condition is True that means the whole statement itself is True
  print("The temperature is bad.")
else:
  print("The temperature is good.")
```

```python
temp = 20
sunny = True

if temp > 0 and temp < 30: # and --> if one of the condition is false that means the whole statement itself is false
  print("The temperature is good.")
else:
  print("The temperature is bad.")

if sunny:
  print("It is sunny outside.")
else:
  print("It is cloudy outside.")
```

```python
# Use of not
if not sunny:
  print("It is cloudy outside.")
else:
  print("It is sunny outside.")
```

```python
is_male = True
is_tall = True

if is_male and is_tall:     # and means both the condition must be true
    print("You are a tall male")
elif is_male and not(is_tall):
    print("You are a short male")
elif not(is_male) and is_tall:
    print("You are not a male but are tall")
else:
    print("You are not a male and not tall")
```

```python
is_male = True
is_tall = True

if is_male or is_tall:  # Or means any of the one condition should be true
    print("You are a male or tall or both")
else:
    print("You neither male nor tall")
```

#### Membership Operators

```python
# Membership Operators = used to test whether a value or variable is found in a sequence
#                        (string, list, tuple, set, or dictionary)
#                        1). in
#                        2). not in
```

```python
# Use of in

word = "APPLE"

letter = input("Guess a letter in the secret word: ").upper()

if letter in word: # If letter is found then in will give answer as True and if not False
  print(f"There ia a {letter}")
else:
  print(f"{letter} was not found")
```

```python
# Use of not in

word = "APPLE"

letter = input("Guess a letter in the secret word: ").upper()

if letter not in word: # If letter is found then in will give answer as True and if not False
  print(f"{letter} was not found")
else:
  print(f"There ia a {letter}")
```

```python
# Use of in
students = {"spongebob", "patric", "sandy"} # This is a sequence

student = input("Enter the name of a student: ").lower()

if student in students:
  print(f"{student} is a student")
else:
  print(f"{student} is not a student")
```

```python
# Use of not in

students = {"spongebob", "patric", "sandy"} # This is a sequence

student = input("Enter the name of a student: ").lower()

if student not in students:
  print(f"{student} is not a student")
else:
  print(f"{student} is a student")
```

```python
grades = {"Sandy" : "A", 
          "Squidward" : "B", 
          "Spongebob" : "C",
          "Patrick" : "D"
          }

student = input("Enter the name of a student: ")

if student in grades:
  print(f"{student} 's grade is {grades[student]}")
else:
  print(f"{student} was not found.")
```

```python
email = "Mani@gmail.com"

if "@" in email and "." in email:
  print("Valid Email")
else:
  print("Not a Valid email.")
```

#### Identity operators


```python
# Compares IDs of the objects

# Identity Operators are:
#   is
#   is not
```

```python
# id() function --> each object has a memory location identified using a unique ID
# ID of an object can be checked using id() function

# Syntax: id(object_name)

str = 'Hello' # Here object is Hello
id(str)

# It does not matter if we store object in a variable or make a object directly both will have same id
```

```python
# is Operator
# Returns True when both the objects have the same ID

x = 10
y = 10
x is y # Will return true as both the variables have same number i.e they will have same id
```

```python
# is not operator

x = "Manikanth"
y = "Aaryan"
x is not y # Will return true as both have different variables
```

```python
# Difference b/w == and is operator

# == is used to compare the values.
# 'is' is used to compare te IDs
```

#### One line If-else

```python
# Cnditional expresssions = A one line shortcut for the if-else statement (ternary operator)
#                           Print or assign one of the two values based on a condition
#                           X if condition else Y

num = 5

print("Positive" if num > 0 else "Negative")
```

```python
num = 6
result = "Even" if num%2 == 0 else "Odd"

print(result)
```

```python
a = 6
b = 7

max_num = a if a > b else b
print(max_num)

min_num = a if a < b else b
print(min_num)
```

```python
age = 17
status = "Adult" if age >= 18 else "Child"
print(status)
```

```python
temp = 32
weather = "Hot" if temp > 25 else "Cold"
print(weather)
```
