#### Some Functions related to strings.

```python
phrase = "Manikanth Here"

print(phrase.upper())
print(phrase.lower())
print(phrase.isupper())

print(len(phrase)) # Tells us how many characters are in a string

print(phrase.replace("Here","is cool")) 
```

```python
print(phrase.upper().isupper()) # Functions can be used together
```

#### Escape Sentence

```python
print(phrase + " is cool ") # Called as Concatenation i.e taking a string and adding another string to it.

print(phrase + " \n is cool ")
print(phrase + " \" is cool \" ")  # \ is an escape staement and \n means to print on a new line
print(phrase + "\\  is cool \" ")
```

```python
print("Manikanth \n Here")
print("Manikanth \" Here \" ")
print("\\ Manikanth Here")
```

#### String Slicing

```python
# Indexing with strings

print(phrase[0])
print(phrase.index("M"))
print(phrase.index("Here")) # Tells us where this word starts!
```

```python
names = "Harry, Shubham"
print(names[0:4])

"""
Here 0 is starting index and 3 is ending index.
Note:- in this 4 is not counted i.e. [0:4] ---> from index position 0 to index
position 3
"""
```

```python
fruit = "Mango"
mangoline1 = len(fruit)
print(mangoline1)
print(fruit[0:4])

print(fruit[:4]) # Agar start index mein kuch nah likha toh its considered 0 by python

print(fruit[:]) # Agar kuch nahi likha toh pura word will be printed or this is same as [0:5]

#Note: () --> used for functions and [] --> slicing
```

```python
# -ve slcing or indexing

print(fruit[0:-3])
print(fruit[-1:len(fruit) -3]) 
# this means ki 4:2 --> doent make any sense

print(fruit[-3:len(fruit)-1])
# this means 2:4
```

#### Strip and Split methods

```python
# str.split always returns a list of strings by removing the character in the parantheses

s = 'abcd efgh ijkl'
s.split(' ') # return a list of strings, based on s, using ' ' as a field separator
```

```python
s = 'abcd:efgh:ijkl'
s.split(':')
```

```python
s = 'abcd:efgh:ijkl'
print(s.split('e'))

s = 'abcd:efgh:ijkl'
print(s.split(':efg'))
```

```python
a = 'this is a very intresting example'
a.split(' ')
```

```python
a = 'this        is a         very intresting        example'
a.split(' ')
```

```python
a.split() # no arguement == None, which really means ANY whitespace, ANY length
```

```python
name = input('Enter Your name: ')
print(f"Hello, {name}!")
```

```python
# str.strip() removes the whitespace from the edges
name.strip()
```

#### Formated Strings/f" Strings

```python
# Formated strings basically let us add the variables in a string.
# To add this variables in a print statement just do f"{variable name}"
```

```python
letter = "Hey my name is {} and i am from {}"
country = 'India'
name = 'Harry'

print(letter.format(country,name))
```

```python
letter = "Hey my name is {1} and i am from {0}"
country = 'India'
name = 'Harry'

print(letter.format(country,name))
```

```python
letter = f"Hey my name is {name} and I am from {country} "
print(letter)

print(f"Hey my name is {name} and I am from {country} ")
```

```python
txt = "For only {price:.2f} dollars" # .2f means just round upto 2 decimal of a float value
print(txt.format(price = 42.1986))
```

```python
price = 39.2345
txt_1 = f"This product is only for {price:.2f} dollars"

print(txt_1)
print(f"For only {price:.2f} dollars")
```

```python
print(type(f"{2*30}"))
print(f"{2*30}")
```

```python
# If we want to see our curly brackets then have double curly brackets

print(f"Hey my name is {{name}} and I am from {{country}} ")
```
