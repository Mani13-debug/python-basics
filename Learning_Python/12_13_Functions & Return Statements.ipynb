#### Variable Scope

```python
# Scope = The region that a variable is recognized
#         A variable is only available from inside the region it is created
#         A global and locally scoped versionsof a variable can be created
```

```python
def display_name():
  name = "Code"  # local scope (available only inside this function)
  print(name)

print(name)  # This will give error as name is a local variable
```

```python
name = "Bro" # Global scope (available inside and outside the functions)

def display_name():
  name = "Code"
  print(name)

display_name()
print(name) # This will print global variable
```

```python
x = 4
print(x)

def hello(): # Local variables destroy themselves once the function is done running
  x = 5
  print(f"The local x is {x}")
  print("Hello Harry")

print(f"The global x is {x}")
hello()
print(f"The global x is {x}")
```

```python
x = 4

def hello():
  x = 5

print(x) # Will print global variable!
```

```python
# global keyword

x = 4 # Global Variable

def hello():
  global x
  x = 5 # local variable

hello()
print(x)
```

```python
# Normal way to find geometric mean

a = 9
b = 8
gmean = (a*b)/(a+b)
print(gmean)

c = 10
d = 25
gmean2 = (c*d)/(c+d)
print(gmean2)
```

```python
# Simple way using functions

def gmean(a, b):
  mean = (a*b)/(a+b)
  print(mean)

a = 9
b = 8
if(a>b):
  print("First number is greater.")
else:
  print("Second number is greater or equal.")

gmean(a , b)

c = 10
d = 25
if(c>d):
  print("First number is greater.")
else:
  print("Second number is greater or equal.")

gmean(c , d)
```

```python
# Efficient way

def gmean(a, b):
  mean = (a*b)/(a+b)
  print(mean)

def isGreater(a , b):
  if(a>b):
    print("First number is greater.")
  else:
    print("Second number is greater or equal.")

a = 10
b = 8

gmean(a , b)
isGreater(a, b)

c = 20
d = 18

gmean(c , d)
isGreater(c, d)

```
