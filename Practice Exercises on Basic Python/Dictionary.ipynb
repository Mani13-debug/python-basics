```python

my_dict = {
  'name' : 'Alice',
  'age': 18,
  'city': 'New York'
}

print(f"Original dictionary: {my_dict}")

# Add a new key value pair

my_dict['profession'] = 'Student'
print(f"Updated dictionary after adding 'profession': {my_dict}")

# Modify Value

my_dict['age'] = 19
print(f"Updated dictionary after modification: {my_dict}")

# Print Key

print('City:', my_dict['city'])

```

Exercise 3

```python
keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]

# Method 1
dictionary = dict(list(zip(keys, values)))
print(dictionary)


# zip function is an iterator that will contain:
# [('a', 1), ('b', 2), ('c', 3)]

# Result: [('a', 1), ('b', 2), ('c', 3)] ---> this is done by list function

# Result: {'a': 1, 'b': 2, 'c': 3} -----> this is done by dict function

"""
dict :- Takes a list of key-value tuples and creates a dictionary out of them.
list :- Converts the zip iterator into a list of tuples.
zip :- The zip() function pairs items from two iterables (a and b) element by element.
"""
```

```python
# Method 2
dictionary = dict(zip(keys, values))
print(dictionary)
```

```python
# Method 3

for i in range(len(keys)):
  dictionary.update({keys[i]:values[i]})
print(dictionary)
```

Exercise 5

```python
dict1 = {'Ten': 10, 'Twenty': 20, 'Thirty': 30}
dict2 = {'Thirty': 30, 'Fourty': 40, 'Fifty': 50}

# Method 1

dict3 = dict1.copy() # Copies the key:value pairs of dict1 but dict3 is a different object
dict3.update(dict2)  # .update() method takes another dictionary (dict2) and adds its key-value pairs to dict3.
print(dict3)

# If a key in dict2 already exists in dict3, the value from dict2 will overwrite the existing value in dict3.
```

```python
# Method 2

dict1 = {'Ten': 10, 'Twenty': 20, 'Thirty': 30}
dict2 = {'Thirty': 30, 'Fourty': 40, 'Fifty': 50}

dict3 = {**dict1, **dict2} # ** is unpacking operator for dictionary
print(dict3)
```

Exercise 12

```python
# Method 1

sample_dict = {
    "name": "Kelly",
    "age": 25,
    "salary": 8000,
    "city": "New york"
}

# Keys to remove
keys = ["name", "salary"]

for k in keys:
  sample_dict.pop(k)

print(sample_dict)
```

```python
# Method 2

sample_dict = {
    "name": "Kelly",
    "age": 25,
    "salary": 8000,
    "city": "New york"
}

# Keys to remove
keys = ["name", "salary"]

sample_dict = {k: sample_dict[k] for k in sample_dict.keys() - keys}
print(sample_dict)
```
